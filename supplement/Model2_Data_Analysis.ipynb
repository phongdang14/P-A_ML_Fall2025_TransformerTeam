```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def histogram(df):
    skip = {'supernova', 'type', 'band', 'n_obs', 'flux_scale', 'redshift'}
    cols = [c for c in df.columns if c not in skip]

    n = len(cols)
    rows = (n + 2) // 3    # number of rows needed
    fig, axes = plt.subplots(rows, 3, figsize=(15, 3*rows))

    axes = axes.flatten()  # flatten so we can index axis[i]

    for i, column in enumerate(cols):
        ax = axes[i]
        df[column].hist(ax=ax, bins=30)
        #ax.set_title(column)
        ax.set_xlabel(column)
        ax.set_ylabel("Count")

    # Hide unused axes if columns not multiple of 3
    for j in range(i+1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.show()
```

Plot histograms of three datasets together

```python
def histogram_all(df1=IIn, df2=II, df3=Ia):
    skip = {'supernova', 'type', 'band', 'n_obs', 'flux_scale', 'redshift'}
    cols = [c for c in df1.columns if c not in skip]

    n = len(cols)
    rows = (n + 2) // 3    # number of rows needed
    fig, axes = plt.subplots(rows, 3, figsize=(15, 3*rows))

    axes = axes.flatten()  # flatten so we can index axis[i]

    for i, column in enumerate(cols):
        ax = axes[i]

        # Plot histograms from all three datasets
        ax.hist(df3[column].dropna(), bins=30, alpha=0.5, label="Ia")
        ax.hist(df1[column].dropna(), bins=30, alpha=0.7, label="IIn")
        ax.hist(df2[column].dropna(), bins=30, alpha=0.5, label="II")


        #ax.set_title(column)
        ax.set_xlabel(column)
        ax.set_ylabel("Count")
        ax.legend()

    # Hide unused axes if columns not multiple of 3
    for j in range(i+1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.show()

histogram_all()
```

Merge three datasets

```python
# Combine them
df_all = pd.concat([IIn, II, Ia], ignore_index=True)

# Plot histogram
histogram(df_all)
```

Clipping data

```python
skip = {'supernova', 'type', 'band', 'n_obs', 'flux_scale', 'redshift'}
feature_cols = [c for c in Ia.columns if c not in skip]


IIn = IIn[feature_cols]
Q1_n = IIn.quantile(0.25)
Q3_n = IIn.quantile(0.75)
IQR_n = Q3_n - Q1_n

lower_n = Q1_n - 1.5 * IQR_n
upper_n = Q3_n + 1.5 * IQR_n

IIn_clip = IIn.clip(lower=lower_n, upper=upper_n, axis=1)


II = II[feature_cols]
Q1 = II.quantile(0.25)
Q3 = II.quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

II_clip = II.clip(lower=lower, upper=upper, axis=1)


Ia = Ia[feature_cols]
Q1_a = Ia.quantile(0.25)
Q3_a = Ia.quantile(0.75)
IQR_a = Q3_a - Q1_a

lower_a = Q1_a - 1.5 * IQR_a
upper_a = Q3_a + 1.5 * IQR_a

Ia_clip = Ia.clip(lower=lower_a, upper=upper_a, axis=1)


histogram_all(IIn_clip, II_clip, Ia_clip)
```
